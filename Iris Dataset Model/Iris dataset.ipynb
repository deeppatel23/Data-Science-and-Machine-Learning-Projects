# IRIS DATASET

This is probably the most versatile, easy and resourceful dataset in pattern recognition literature. Nothing could be simpler than the Iris dataset to learn classification techniques.

<b>Predict class of flower on basis of given attributes.</b>

```python
#import important libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('irisdata.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
unique = set(df['Species'])
unique
```

```python
#transforming species name to digits
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df.Species = encode.fit_transform(df.Species)
df
```

```python
#train-test-split
y_data = df['Species']
x_data=df.drop('Species',axis=1) # all data except species
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.15, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
#Create the linear regression object
lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
test_y_hat = lm.predict(x_test)
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - y_test) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , y_test) )
```
